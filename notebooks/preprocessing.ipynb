```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the input image
image = cv2.imread("../data/samples/ecg.jpeg")
```

```python
plt.imshow(image, cmap='gray')
```

```python
# Step 1: Resize (optional, assuming it's already 916 x 1280)
image = cv2.resize(image, (1280, 916))
plt.imshow(image, cmap='gray')
```

```python
# Step 2: Convert to Grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Invert grayscale image (so black trace becomes white)
inverted = 255 - gray

# Apply CLAHE on inverted image
clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
enhanced = clahe.apply(inverted)

# Invert back
enhanced = 255 - enhanced
plt.imshow(enhanced, cmap='gray')
```

```python
# Step 6: Save the final preprocessed image
cv2.imwrite("../data/preprocessed/ecg_preprocessed1.jpg", enhanced)
```
